````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scipy.stats as stats
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/diamonds.csv')
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"

#code here
sns.set()

with sns.axes_style('white'):
    sns.scatterplot(data=df, x='carat', y='price', hue='clarity', palette=palette, size='depth', sizes=(5, 10), hue_order=['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'], linewidth=0.1)

plt.show()
````

````python
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)

with sns.axes_style("dark"):
    g = sns.FacetGrid(df, hue="cut", col="cut", height=4)
g.map(hexbin, "carat", "price", extent=[0, 3, 0, 20000]);
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)

#code here
sns.set_style('ticks')
sns.pairplot(df, hue='cut', corner=True, x_vars=['carat', 'depth', 'table', 'price'], y_vars=['carat', 'depth', 'table', 'price'], markers=['+', 'o', 's', 'D', 'x'])

plt.show()
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here

sns.set()

with sns.axes_style('white'):
    sns.jointplot('carat', 'price', df, kind='kde')

plt.show()
````

````python
#code here

sns.set()

with sns.axes_style('ticks'):
    sns.jointplot('carat', 'price', df, kind='reg', stat_func=stats.pearsonr, xlim=(0, 2.5), ylim=(0, 20000))

plt.show()
````
