Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot()

ax.plot(x, cos, color='#ff4141', label='Cos X')
ax.plot(x, cos3, linestyle='dashed', color='#2e972e', label='3 Cos X')
ax.plot(x, sin, linestyle='dashdot', color='#3f3fff', label='Sin (X - 90)')
ax.plot(x, sin2, marker='x', linestyle='None', color='#d6d65b', label='2 Sin X')

ax.legend(loc='center', prop={'size': 7})
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')

plt.show()
```
