1. Machine Learning merupakan cabang dari Artificial Intelligence (AI) dengan kemampuan mesin untuk mengakses data yang ada dengan perintah mereka sendiri. Machine Learning juga mampu mempelajari data yang ada dan melakukan tugas-tugas tertentu. Machine Learning mampu melakukan ini dengan metode mempelajari algoritma dan model statistik yang ada.


2. Data feature adalah variabel yang memprediksi suatu kejadian/target. Sedangkan data target adalah hasil dari prediksi kejadian tersebut.


3. Supervised Learning adalah algoritma machine learning yang dapat menerapkan informasi yang telah ada pada data dengan memberikan label tertentu (data latih), misalnya data yang telah diklasifikasikan sebelumnya (terarah). Sedangkan Unsupervised Learning adalah algoritma machine learning yang digunakan pada data yang tidak mempunyai informasi yang dapat diterapkan secara langsung (tidak terarah). Algoritma ini diharapkan mampu menemukan struktur tersembunyi pada data yang tidak berlabel.

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

- n features : 1
  
Tree Graph :

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

model_dtr = DecisionTreeRegressor(max_depth = 6, max_leaf_nodes = 7, min_samples_leaf = 5, min_samples_split = 2)

x_matriks = x[:, np.newaxis]
x_test = x_new[:, np.newaxis]

model_dtr.fit(x_matriks, y)

y_test = model_dtr.predict(x_test)
y_train = model_dtr.predict(x_matriks)

fig, ax = plt.subplots(figsize = (12, 6))
ax.scatter(x, y, color = 'blue', label = 'Data Asli')
ax.scatter(x, y_train, color = 'orange', label = 'Prediksi terhadap Data Training')
ax.plot(x_new, y_test, color = 'green', label = 'Prediksi terhadap Data Baru')

fig.legend(loc = 'upper right')
plt.show()
```
