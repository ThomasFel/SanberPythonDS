```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2 * -np.pi, 2 * np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, squeeze=False, sharex=True, sharey=True, figsize=(20, 8))

ax[0][0].plot(x, tan, color='blue', label='Tan X')
ax[0][0].set_title('Grafik Tan X')

ax[0][1].plot(x, sin, color='red', label='Sin X')
ax[0][1].set_title('Grafik Sin X')

ax[0][2].plot(x, cos, color='green', label='Cos X')
ax[0][2].set_title('Grafik Cos X')

fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2 * -np.pi, 2 * np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, squeeze=False, sharex=True, sharey=True, figsize=(13, 8))

ax[0][0].plot(x, y, linewidth=1.5)
ax[0][0].set_title('Grafik Cos X')

ins1 = ax[0][0].inset_axes([-0.1, 0.9, 0.3, 0.3])
ins1.plot(x, y5, color='#d6d65b', linewidth=1.5)
ins1.plot(x, y, linewidth=1.5)
ins1.set_title('Grafik Cos X\u2075')

ins2 = ax[0][0].inset_axes([0.9, 0.9, 0.3, 0.3])
ins2.plot(x, y4, color='black', linewidth=1.5)
ins2.plot(x, y, linewidth=1)
ins2.set_title('Grafik Cos X\u2074')

ins3 = ax[0][0].inset_axes([-0.1, -0.1, 0.3, 0.3])
ins3.plot(x, y2, color='red', linewidth=1.5)
ins3.plot(x, y, linewidth=1.5)
ins3.set_title('Grafik Cos X\u00b2')

ins4 = ax[0][0].inset_axes([0.9, -0.1, 0.3, 0.3])
ins4.plot(x, y3, color='green', linewidth=1.5)
ins4.plot(x, y, linewidth=1.5)
ins4.set_title('Grafik Cos X\u00b3')


plt.show()
```
